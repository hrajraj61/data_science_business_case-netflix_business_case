## **3.1 DataType Conversion**
> In this section, we will perform data type conversions on the columns of the Netflix DataFrame to ensure that each column's data type accurately reflects the nature of the data it contains. This will help optimize memory usage and improve the efficiency of any analysis we conduct. Hereâ€™s a breakdown of our approach for each column:

| **S.No** | **Column**       | **Current Type** | **Desired Type** | **Reason**                                                                                  |
|----------|-------------------|------------------|------------------|---------------------------------------------------------------------------------------------|
| 1        | **show_id**       | object           | string           | Unique identifier for each show; preserving leading zeros or special characters.           |
| 2        | **type**          | object           | category         | Limited number of unique values; saves memory and enhances performance.                     |
| 3        | **title**         | object           | string           | The title of each show is textual data and should be treated as a string.                  |
| 4        | **director**      | object           | string           | Directors' names are textual data; should be stored as strings for full representation.    |
| 5        | **cast**          | object           | string           | Cast names should be represented as strings to accommodate full names without truncation.   |
| 6        | **country**       | object           | category         | Limited number of unique values; ideal for conversion to a categorical type.               |
| 7        | **date_added**    | object           | datetime         | The date when the show was added should be in datetime format for time-based analysis.     |
| 8        | **release_year**  | object           | integer          | Numeric value; should be stored as an integer for accurate mathematical operations.         |
| 9        | **rating**        | object           | category         | Ratings have a limited set of values and should be stored as categorical data.             |
| 10       | **duration**      | object           | integer          | Duration is numeric (in minutes or seasons); we will extract the integer part for accuracy.|
| 11       | **listed_in**     | object           | category         | Describes genres, having a limited set of values; appropriate for categorical type.        |
| 12       | **description**   | object           | string           | Descriptions are textual and should be represented as strings for clarity and accuracy.    |

> By converting these columns to their appropriate data types, we aim to ensure the integrity and efficiency of our data processing and analysis.


```python
# Converting datatypes

# Column: show_id (This should be a string)
netflix_dataframe['show_id'] = netflix_dataframe['show_id'].astype(str)

# Column: type (This should be a category)
netflix_dataframe['type'] = netflix_dataframe['type'].astype('category')

# Column: title (This should be a string)
netflix_dataframe['title'] = netflix_dataframe['title'].astype(str)

# Column: director (This should be a string)
netflix_dataframe['director'] = netflix_dataframe['director'].astype(pd.StringDtype())

# Column: cast (this should be a string)
netflix_dataframe['cast'] = netflix_dataframe['cast'].astype(pd.StringDtype())

# Column: country (this should be a category)
netflix_dataframe['country'] = netflix_dataframe['country'].astype('category')

# Column: date_added (this should be a datetime)
netflix_dataframe['date_added'] = pd.to_datetime(netflix_dataframe['date_added'].str.strip())

# Column: release_year (this should be a integer)
netflix_dataframe['release_year'] = netflix_dataframe['release_year'].astype(int)

# Column: rating (this should be a category)
netflix_dataframe['rating'] = netflix_dataframe['rating'].astype('category')

# Column: duration (this should be a integer)
# keeping only integer part (e.g: 91 minutes -> 91, 2 season : 2)
netflix_dataframe['duration'] = netflix_dataframe['duration'].apply(lambda value: int(value.split(' ')[0]) if not pd.isna(value) else None)

# Column: listed_in (this should be a category)
netflix_dataframe['listed_in'] = netflix_dataframe['listed_in'].astype('category')

# Column: description (this should be a string)
netflix_dataframe['description'] = netflix_dataframe['description'].astype(str)
```

```python
# Glance at data now

print(netflix_dataframe.dtypes)
netflix_dataframe.head()
```

## **3.2 Exploding Columns**
> In this section, we will transform the `netflix_dataframe` by exploding specific columns to separate the values into individual rows.

1. **Exploding the `cast` column:** We first split the `cast` column on commas and then explode the resulting lists to create a new row for each actor in the cast. The index is reset and the column is renamed to `actor`.
   
2. **Exploding the `listed_in` column:** Similarly, we split the `listed_in` column on commas, explode it, and rename it to `genre` after stripping any whitespace.
   
3. **Exploding the `country` column:** We perform the same operation for the `country` column, ensuring each country is represented in a separate row.

3. **Exploding the `director` column:** We perform the same operation for the `director` column, ensuring each director is represented in a separate row.

> Finally, we clean up any leading or trailing whitespace in the `actor`, `genre`, `country` and `director` columns. The resulting DataFrame will be displayed, showcasing the transformed data.

```python

# Update DataFrame with exploded cast
netflix_dataframe = netflix_dataframe.assign(cast=netflix_dataframe.cast.str.split(',')).explode('cast')

# Reset index and rename cast
netflix_dataframe = netflix_dataframe.reset_index(drop=True).rename(columns={'cast': 'actor'})

#  Update DataFrame with exploded listed_in
netflix_dataframe = netflix_dataframe.assign(listed_in=netflix_dataframe.listed_in.str.split(',')).explode('listed_in')

# Strip whitespace from the splitted actor values
netflix_dataframe['actor'] = netflix_dataframe['actor'].str.strip()

# Reset index and rename listed_in
netflix_dataframe = netflix_dataframe.reset_index(drop=True).rename(columns={'listed_in': 'genre'})

# Strip whitespace from the splitted genre values
netflix_dataframe['genre'] = netflix_dataframe['genre'].str.strip()

#  Update DataFrame with exploded country
netflix_dataframe = netflix_dataframe.assign(country=netflix_dataframe.country.str.split(',')).explode('country')

# Strip whitespace from the splitted country values
netflix_dataframe['country'] = netflix_dataframe['country'].str.strip()

# Update Datframe with exploded director
netflix_dataframe = netflix_dataframe.assign(director=netflix_dataframe.director.str.split(',')).explode('director')

# Strip whitespace from the splitted director values
netflix_dataframe['director'] = netflix_dataframe['director'].str.strip()

# Display
netflix_dataframe.head()
```

### **3.3.1 Handling `duration`**
> There are **3 missing values** in the **duration** column. We observed that for some shows, the duration is mentioned within the **rating column**. We can utilize this information to impute the **missing duration** values.



```python
# display show_id where duration is na

missing_duration_dataframe = netflix_dataframe[netflix_dataframe['duration'].isna()]
show_id_of_missing_duration = missing_duration_dataframe['show_id']
missing_duration_dataframe.head(3)
```

```python
# Filling na value for `duration`

import pandas as pd
def fill_duration(row):
  if pd.isna(row['duration']):
    if isinstance(row['rating'], str) and 'min' in row['rating']:
      try:
        return int(row['rating'].split(' ')[0])
      except ValueError:
        pass
  return row['duration']

netflix_dataframe['duration'] = netflix_dataframe.apply(fill_duration, axis=1)

netflix_dataframe['duration'] = netflix_dataframe.groupby('type', observed=False)['duration'].transform(lambda x: x.fillna(x.mean()))
```

```python
# print row where show_id is in show_id_of_missing_duration

netflix_dataframe[netflix_dataframe['show_id'].isin(show_id_of_missing_duration)].head(3)
```

### **3.3.2 Handling `rating`**
> By careful consideration. It is better to fillup the **missing** value to **`unrated`** . As we are not sure what we can use for **missing rating**.

```python
# display show_id where rating is na

missing_rating_dataframe = netflix_dataframe[netflix_dataframe['rating'].isna()]
show_id_of_missing_rating = missing_rating_dataframe['show_id']
missing_rating_dataframe.head(3)
```

```python

# Add 'unrated' to the categories of the 'rating' column
netflix_dataframe['rating'] = netflix_dataframe['rating'].cat.add_categories('unrated')

# Fill `na` value with unrated
netflix_dataframe['rating'] = netflix_dataframe['rating'].fillna("unrated")

netflix_dataframe[netflix_dataframe['show_id'].isin(show_id_of_missing_rating)].head(3)
```

### **3.3.3 Handling `date_added`**

> In our Netflix dataset, we aim to address **`missing values`** in the **`date_added`** column by leveraging the **average difference between the `date_added` and `released_year`** for all **non null shows**. and add the **difference** to **`release_year`**.


```python
# Finding out the missing date added
missing_date_added_dataframe = netflix_dataframe[netflix_dataframe['date_added'].isna()]
show_id_of_missing_date_added = missing_date_added_dataframe['show_id']

missing_date_added_dataframe.head(3)
```

```python

# dataframe with dropna for column date_added
netflix_dataframe_dropna = netflix_dataframe.dropna(subset=['date_added'])

# Calculate the average difference between date_added year and release_year
avg_year_diff = (netflix_dataframe_dropna['date_added'].dt.year - netflix_dataframe_dropna['release_year']).mean()

# Fill missing date_added values with release_year + avg_year_diff
netflix_dataframe['date_added'] = netflix_dataframe['date_added'].fillna(pd.to_datetime(netflix_dataframe['release_year'], format='%Y') + pd.DateOffset(years=int(avg_year_diff)))
```

```python

netflix_dataframe[netflix_dataframe['show_id'].isin(show_id_of_missing_date_added)].head(3)


```

### **3.3.4 Handling `country`**

> To handle **missing `country`** value we will use **most frequent country** to the missing data

```python
# Getting show_id for missing country value

missing_country_dataframe = netflix_dataframe[netflix_dataframe['country'].isna()]
show_id_of_missing_country = missing_country_dataframe['show_id']

missing_country_dataframe.head(3)
```

```python
# finding out most_common country to be filled for na value

# Group by show_id and get unique countries for each show
unique_countries_per_show = netflix_dataframe.groupby('show_id')['country'].unique()

# Printing unique countries
print("########## Unique Country per Show #########\n", unique_countries_per_show)

# Finding out most common country
most_common_country = unique_countries_per_show.explode().dropna().mode()

print("\nMost Common Country:", most_common_country.values[0])
```

```python
# fillna for country with most_common_country

netflix_dataframe['country'] = netflix_dataframe['country'].fillna(most_common_country[0])

netflix_dataframe[netflix_dataframe['show_id'].isin(show_id_of_missing_country)].head(3)
```

### **4.2.5 Netflix Age Breakdown Globally** ðŸ‘¶ðŸ§’ðŸ‘©â€ðŸ¦±ðŸ‘´
> Let's try to breakdown how Netflix audience differ from country by country

```python
# Get the top 10 countries based on the number of unique shows available
# 'nunique()' counts unique 'show_id's in each country
# Sort countries by descending order of unique show counts
# Reset the index and extract only the 'country' column
country_to_consider = (
    netflix_dataframe
    .groupby('country')['show_id']
    .nunique()
    .sort_values(ascending=False)[:10]
    .reset_index()['country']
)

# Filter the dataframe to include only the top 10 countries
# Group by 'country' and 'rating' to count unique shows per rating
df = (
    netflix_dataframe[netflix_dataframe['country'].isin(country_to_consider)]
    .groupby(['country', 'rating'])['show_id']
    .nunique()
    .reset_index()
)

# Remove rows where the count of 'show_id' is 0
df = df[df["show_id"] != 0]

# Add a new column 'age_group' by applying the 'age_group' function to the 'rating' column
df["age_group"] = df['rating'].apply(age_group)

# Group by 'country' and 'age_group' and sum the number of shows per age group
# Sort the resulting dataframe by 'show_id' in descending order
df = (
    df.groupby(['country', 'age_group'])['show_id']
    .sum()
    .reset_index()
    .sort_values('show_id', ascending=False)
)

# Function to calculate the percentage of shows for each country
def percent(x):
    # Calculate the percentage of 'show_id' for each row relative to the total for the country
    x['percent'] = round(x['show_id'] / x['show_id'].sum(), 2)
    return x

# Apply the 'percent' function to each country group and calculate the percentage of shows
df = df.groupby('country').apply(lambda x: percent(x))

# Display the first few rows of the dataframe
df.head()
```

```python
# Create a pivot table to reshape the dataframe for a heatmap
# Rows will be 'age_group', columns will be 'country', and values will be the 'percent'
df_heatmap = df.pivot(index='age_group', columns='country', values='percent')

# Filter the pivot table to display only the specified age groups in the given order
# 'country_to_consider' is already the list of top 10 countries, so the table is limited to those columns
df_heatmap.loc[['Little Kids', 'Older Kids', 'Teens', 'Adults'], country_to_consider]
```

```python
# Set up the plot with a specified figure size
fig, ax = plt.subplots(figsize=(12, 12))

# Create a heatmap using seaborn with custom settings
sns.heatmap(
    data=df_heatmap,               # Data for the heatmap (percent values for age group by country)
    ax=ax,                         # Axis to plot on
    cmap=netflix_palette,          # Color palette for the heatmap (assuming 'netflix_palette' is predefined)
    square=True,                   # Set each cell to be square-shaped
    linewidth=2.5,                 # Set the width of the lines separating the cells
    cbar=False,                    # Disable the color bar on the side
    annot=True,                    # Annotate each cell with its value
    fmt='.0%',                     # Format annotations as percentages with no decimal places
)

# Customize the y-axis labels (age groups) to be serif font and no rotation
ax.set_yticklabels(ax.get_yticklabels(), fontfamily='serif', rotation=0)

# Customize the x-axis labels (countries) to be serif font with a 45-degree rotation for better readability
ax.set_xticklabels(ax.get_xticklabels(), fontfamily='serif', rotation=45)

# Remove axis labels for a cleaner look
ax.set_xlabel('')
ax.set_ylabel('')

# Remove tick marks on both axes but keep the labels visible
ax.tick_params(axis='both', length=0)

# Set the title for the heatmap
ax.set_title('Target Ages By Country', fontfamily='serif', fontsize=16)

# Display the plot
plt.show()
```

```python
def get_month_day_added_df(df, typ):
    df = df[df['type'] == typ][['title', 'date_added']].drop_duplicates()
    df['month_added'] = df['date_added'].dt.month
    df['day_added'] = df['date_added'].dt.day
    return df



released_tv = get_month_day_added_df(netflix_dataframe, "TV Show")
released_movie = get_month_day_added_df(netflix_dataframe, "Movie")



released_tv_month_count = released_tv.groupby('month_added')['title'].count().reset_index()
released_movie_month_count = released_movie.groupby('month_added')['title'].count().reset_index()

released_tv_day_count = released_tv.groupby('day_added')['title'].count().reset_index()
released_movie_day_count = released_movie.groupby('day_added')['title'].count().reset_index()
```
